```python
import scipy.io
import numpy as np
import math
from sklearn import preprocessing
import random
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
%matplotlib inline 

old_err_state = np.seterr(divide='raise')
ignored_states = np.seterr(**old_err_state)

class Neural_Network(object):
    def __init__(self):        
        #Define Hyperparameters
        self.inputLayerSize = 785
        self.outputLayerSize = 10
        self.hiddenLayerSize = 201
        
        #Weights (parameters)
        self.W1 = 0.01 * np.random.randn(self.inputLayerSize,self.hiddenLayerSize)       #785*201
        self.W2 = 0.01 * np.random.randn(self.hiddenLayerSize,self.outputLayerSize)      #201*10
        
    def forward(self, x):
        #Propogate inputs though network
        self.s = np.dot(self.W1.T, x)       #201*1
        self.y = np.tanh(self.s)            #201*1
        self.z = np.dot(self.W2.T, self.y)  #10*1
        g = self.sigmoid(self.z)     
        
        return g
    
    def sigmoid(self, z):
        #Apply sigmoid activation function to scalar, vector, or matrix
        return 1/(1+np.exp(-z))             #10*1
    
    def sigmoidPrime(self,z):
        #Gradient of sigmoid
        d = 1+np.exp(-z)                    #10*1
        sqr =np.multiply(d, d)          
        r = np.exp(-z)/sqr                  #10*1
        return r
    
    def costFunction(self, x, label):
        #Compute cost for given X,y, use weights already stored in class.
        self.g = self.forward(x)            #10*1
        d = - np.dot(label.T, np.log(self.g)) - np.dot((1 - label).T, np.log(1 - self.g))
        J = 0.5*np.dot(d.T,d)
        return J
        
    def costFunctionPrime(self, x, label):
        #Compute derivative with respect to W and W2 for a given X and y:
        self.g = self.forward(x)            #10*1
        dL = np.multiply((1-label), 1/(1-self.g)) - np.multiply(label, 1/self.g)
        
#         print label.shape, self.g.shape   #10*1 10*1
        delta3 = np.multiply(dL, self.sigmoidPrime(self.z))
#         print "delta3 ", delta3.shape     #10*1
        #201 * 10       201*1
        dJdW2 = np.dot(self.y, delta3.T)

        t = np.tanh(self.s)   #
        # print t.shape, delta3.shape #201*1   10*1
        tan = (1 - np.multiply(t, t))
        delta2 = np.multiply(np.dot(self.W2, delta3), tan)

        dJdW1 = np.dot(x, delta2.T)
        # print dJdW1
        return dJdW1, dJdW2
    
    def vali(self, xv, yx):
        l_predict = self.forward(xv.T).T
        result = 0
        for i in range(len(yx)):
            pre = l_predict[i].tolist()[0]
            # print pre
            idx = pre.index(max(pre))
            # print idx, yx[i]
            if yx[i][idx] == 1:
                result += 1
        return 1.0 * result/len(yx)
```

```python
data = scipy.io.loadmat('dataset/mnist.mat')
print data.keys()
train_labels = data['trainY']
train_img = data['trainX']
t = np.ones((train_img.shape[0], train_img.shape[1] + 1))
t[:,:-1] = train_img
train_img = t
print len(train_img[0])
print train_img.shape
lb = preprocessing.LabelBinarizer()
lb.fit(range(10))
train_labels = lb.transform(train_labels[0])

test_img = data['testX']
test_labels = data['testY']
t = np.ones((test_img.shape[0], test_img.shape[1] + 1))
t[:,:-1] = test_img
test_img = t
print test_img.shape
print test_labels


train_img ,train_labels = shuffle(train_img, train_labels, random_state=0)
train_img = preprocessing.scale(train_img)
X = train_img[:55000]
y = train_labels[:55000]
Xv = train_img[55000:]
print Xv.shape
yv = train_labels[55000:]
```

```python
learn_rate = 0.01
decay = 0.9
interval = 1000
NN = Neural_Network()
train_acc = []
vali_acc = []
train_error = []
test_acc = []


for i in range(200):
    cost = 0
    for j in range(10000):
        idx = random.randint(0, len(X)-1)
#     for j in range(len(X)):
#         idx = j
        x1 = np.matrix(X[idx]).T
        y1 = np.matrix(y[idx]).T
        cost1 = NN.costFunction(x1,y1)
        cost1 = cost1.max()
        cost += cost1
        dJdW1, dJdW2 = NN.costFunctionPrime(x1,y1)
        NN.W1 = NN.W1 - learn_rate*dJdW1
        NN.W2 = NN.W2 - learn_rate*dJdW2
        if (j + 1) % interval == 0:
            er1 = NN.vali(X, y)
            er2 = NN.vali(Xv, yv)
            train_acc.append(er1)
            vali_acc.append(er2)
            train_error.append(cost)
            cost_pre = cost
            cost = 0

#             print i, j + 1, er1, er2, cost_pre
    
    l_predict = NN.forward(test_img.T).T
    count = 0
    for j in range(len(l_predict)):
        pre = l_predict[j].tolist()[0]
        idx = pre.index(max(pre))
        if idx == test_labels[0][j]:
            count += 1
    acc = 1.0*count/len(l_predict)
    test_acc.append(acc)
    
    print i, er1, er2, cost_pre,acc
    
    if (i + 1) % 10 == 0:
        learn_rate = learn_rate * decay
```

```python
plt.plot([i * interval for i in range(len(train_acc))], train_acc)
plt.plot([i * interval for i in range(len(vali_acc))], vali_acc)
plt.legend(['training accuracy', 'validation accuracy'], loc = 'upper right')
plt.show()

plt.plot([i * interval for i in range(len(train_error))], train_error)
plt.legend(['training error'], loc = 'upper right')
plt.show()

plt.plot([i*interval for i in range(len(test_acc))], test_acc)
plt.legend(['test accuracy'], loc = 'upper right')
plt.show()
```
